```python
from itertools import product

import numpy as np
import matplotlib.pyplot as plt

plt.rc('font', size=11)
```

```python
def load_data(filename):
    return np.loadtxt(filename).T.reshape((6, 10, 50000))

psi_frw_data = load_data('data/psi_frw.dat')
phi_frw_data = load_data('data/phi_frw.dat')
psi_bck_data = load_data('data/psi_bck.dat')
phi_bck_data = load_data('data/phi_bck.dat')
```

```python
np.min(np.hstack([psi_frw_data, psi_bck_data])), np.max(np.hstack([psi_frw_data, psi_bck_data]))
```

```python
np.min(np.hstack([phi_frw_data, phi_bck_data])), np.max(np.hstack([phi_frw_data, phi_bck_data]))
```

```python
x = np.arange(psi_frw_data.shape[2])

fig, axes = plt.subplots(6, 10, figsize=(12, 7))
fig.subplots_adjust(wspace=0., hspace=0.)
for i, j in product(range(6), range(10)):
    ax = axes[i, j]
    ax.set_xlim([0, 50000])
    ax.set_ylim([-230, 70])
    x_ticks = np.arange(0, 50001, 12500) if i == 5 else []
    xticklabels = np.arange(0, 101, 25) if i == 5 else []
    if j < 9:
        x_ticks = x_ticks[:-1]
        xticklabels = xticklabels[:-1]
    y_ticks = np.arange(-200, 51, 50) if j == 0 else []
    ax.set_xticks(x_ticks, xticklabels, rotation=60)
    # ax.set_xticklabels(xticklabels)
    ax.set_yticks(y_ticks)
    ax.plot(x, psi_frw_data[i, j], lw=0.5, c='blue')
    ax.plot(x, psi_bck_data[i, j], lw=0.5, c='red')
fig.supxlabel('t (ps)')
fig.supylabel('$\Psi$ ($\degree$)', x=0.06)
plt.savefig('FigS5_1.png', dpi=300, bbox_inches='tight')
plt.show()
```
